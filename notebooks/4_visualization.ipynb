```python
# Importation des bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le fichier de dataset weatherAUS.csv
dataset_path = "../data/weatherAUS.csv"  # Chemin vers votre dataset
data = pd.read_csv(dataset_path)

# Afficher les premières lignes pour vérifier les données
print("Aperçu des données :")
display(data.head())

# Résumé des informations générales
print("\nInformations générales sur le dataset :")
data.info()

# Statistiques descriptives des variables numériques
print("\nStatistiques descriptives :")
display(data.describe())

# Compter les valeurs manquantes par colonne
print("\nValeurs manquantes par colonne :")
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])

# Visualisation de la répartition des valeurs manquantes
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Répartition des valeurs manquantes dans le dataset")
plt.show()

# Visualisation des variables catégorielles (exemple : 'Location')
plt.figure(figsize=(12, 6))
sns.countplot(y="Location", data=data, order=data["Location"].value_counts().index)
plt.title("Distribution des observations par Location")
plt.show()

# Visualisation d'une variable numérique (exemple : MinTemp)
plt.figure(figsize=(10, 6))
sns.histplot(data['MinTemp'].dropna(), kde=True, color="skyblue")
plt.title("Distribution de MinTemp")
plt.xlabel("MinTemp")
plt.show()
```
